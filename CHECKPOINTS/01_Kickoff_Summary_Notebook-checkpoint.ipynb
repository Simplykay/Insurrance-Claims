## 02. Data Tables Overview

### Claims Table  
Each row represents a single insurance claim, including identifiers, claim type, dates, initial estimates, ultimate amounts, accident characteristics, and links to policyholders and third parties. This table forms the core dataset for modelling.

### Policyholder Table  
Each row represents an insured customer and includes demographic details, policy information, coverage type, and historical risk indicators. It links to the claims table via policy or customer identifiers.

### Third Party Table  
Each row represents a third party involved in a claim, such as another driver or injured person. It captures involvement details, liability indicators, and severity-related attributes critical to understanding claim costs.


## 03. Key Insurance Definitions

### First Notice of Loss (FNOL)  
FNOL is the moment when the insurer is first informed that a loss or accident has occurred. At this stage, information is incomplete, yet insurers must begin claims handling and reserve setting.

### Estimated Claim Amount  
The estimated claim amount is the insurer’s early assessment of the expected cost of a claim, used to establish financial reserves and guide initial claim handling decisions.

### Ultimate Claim Amount  
The ultimate claim amount is the final total cost of a claim after full settlement, including all payments, expenses, and recoveries.

### Third-party Severity  
Third-party severity describes the financial impact of claims involving non-policyholders. High severity typically arises from serious injuries, multiple claimants, or complex legal liability.


## 04. Expected Claim Cost Distribution

Insurance claim costs usually follow a **right-skewed distribution**, where most claims are low value and a small number of claims account for very large losses. This pattern underpins the need for robust reserving and predictive modelling.


### Plot a sample claim cost histogram

```python
# importing relevant libraries for creating histogram
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Right-skewed claim costs 
claim_costs = np.random.lognormal(mean=9, sigma=1.1, size=2000)

plt.figure(figsize=(8,4))
plt.hist(claim_costs, bins=50)
plt.xlabel("Claim Cost")
plt.ylabel("Number of Claims")
plt.title("Sample Claim Cost Distribution")
plt.tight_layout()
plt.show()
```
