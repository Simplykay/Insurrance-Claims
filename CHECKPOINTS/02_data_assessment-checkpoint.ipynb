## 1. Load Tables

```python
# Importing relevant libraries
import pandas as pd
import numpy as np

# creating document paths for loading
claims_path = "claims.csv"
policyholders_path = "policyholders.csv"
third_parties_path = "third_parties.csv"

 
import os
if not os.path.exists(claims_path) and os.path.exists("/mnt/data/claims.csv"):
    claims_path = "/mnt/data/claims.csv"
if not os.path.exists(policyholders_path) and os.path.exists("/mnt/data/policyholders.csv"):
    policyholders_path = "/mnt/data/policyholders.csv"
if not os.path.exists(third_parties_path) and os.path.exists("/mnt/data/third_parties.csv"):
    third_parties_path = "/mnt/data/third_parties.csv"
    
# assigning uploaded documents to variables
claims = pd.read_csv(claims_path)
policyholders = pd.read_csv(policyholders_path)
third_parties = pd.read_csv(third_parties_path)

print("All documents uploaded and assigned successfully")
claims.shape, policyholders.shape, third_parties.shape
```

## 2. Initial Inspection (`head`, `info`, `describe`)

```python
# importing relevant library
import pandas as pd

def overview(df: pd.DataFrame, name: str, n: int = 5):
    print(f"\n{'='*80}\n{name} — head({n})\n{'='*80}")
    display(df.head(n))
    
    print(f"\n{'='*80}\n{name} — info()\n{'='*80}")
    df.info()
    
    # describe() for numeric and object columns
    print(f"\n{'='*80}\n{name} — describe(include='all')\n{'='*80}")
    display(df.describe(include='all'))

overview(claims, "Claims")
overview(policyholders, "Policyholders")
overview(third_parties, "Third Parties")
```

## 3. Missing Values per Column

```python
def missing_table(df: pd.DataFrame, name: str) -> pd.DataFrame:
    out = df.isna().sum().to_frame("missing_count")
    out["missing_pct"] = (out["missing_count"] / len(df) * 100).round(2)
    out = out.sort_values(["missing_count", "missing_pct"], ascending=False)
    out.index.name = "column"
    out.reset_index(inplace=True)
    out.insert(0, "table", name)
    return out

missing_all = pd.concat([
    missing_table(claims, "claims"),
    missing_table(policyholders, "policyholders"),
    missing_table(third_parties, "third_parties"),
], ignore_index=True)

missing_all
```

## _Insight:_ 
- **Missing values in the _Ultimate_claim_Amount_ column are primarily attributable to open or unresolved claims for which final settlement has not yet occurred**
- **The absence of settlement dates is consistent with claims that remain in progress. This aligns with the missing ultimate claim amounts and indicates that missingness is driven by claim status rather than incomplete data capture**


## 4. Target Variable Distribution

Target: **Ultimate_Claim_Amount** (Claims table)


```python
# importing relevant libraries
import matplotlib.pyplot as plt

target_col = "Ultimate_Claim_Amount"

# Basic stats + skewness
target = claims[target_col].dropna()
print("Target count:", target.shape[0])
print("Target summary:")
display(target.describe())

skewness = target.skew()
print("Skewness:", round(float(skewness), 4))

# Histogram
plt.figure(figsize=(9, 4))
plt.hist(target, bins=50)
plt.xlabel(target_col)
plt.ylabel("Number of claims")
plt.title("Distribution of Ultimate Claim Amount")
plt.tight_layout()
plt.show()

# Log-scale view (often helpful for skewed loss data)
plt.figure(figsize=(9, 4))
plt.hist(np.log1p(target), bins=50)
plt.xlabel(f"log1p({target_col})")
plt.ylabel("Number of claims")
plt.title("Distribution of log-transformed Ultimate Claim Amount")
plt.tight_layout()
plt.show()
```

## **__Insight:__** 
**The ultimate claim amount displays a classic insurance severity profile, characterised by frequent low-cost claims and a heavy-tailed distribution of infrequent but high-severity losses. These tail losses are the primary drivers of reserve volatility and capital risk, underscoring the importance of severity-aware modelling and conservative reserving assumptions**

## 5. Extreme Outliers

We inspect the largest claims to identify potential extreme outliers.


```python
top_n = 15
display(claims[[ "Claim_ID", "Policy_ID", "Claim_Type", "Estimated_Claim_Amount", "Ultimate_Claim_Amount", "Severity_Band", "Status"]]
        .sort_values("Ultimate_Claim_Amount", ascending=False)
        .head(top_n))

# Simple IQR-based flag 
q1, q3 = target.quantile([0.25, 0.75])
iqr = q3 - q1
upper_fence = q3 + 1.5 * iqr
outlier_rate = (target > upper_fence).mean() * 100

print("IQR upper fence:", float(upper_fence))
print("Outlier rate (target > upper fence):", round(float(outlier_rate), 2), "%")
```

## **__Insight:__**
**The extreme values identified in the ultimate claim amount represent a low-frequency, high-severity segment of the portfolio, predominantly associated with catastrophic fire and theft claims. Although these observations constitute a minority of total claims (approximately 13% above the IQR upper fence), they account for a disproportionate share of aggregate losses and therefore drive reserve volatility and capital risk. These claims reflect genuine severity exposure rather than data anomalies and should be retained for modelling and reserving. From an actuarial perspective, the results underscore the importance of explicitly modelling severity tails and applying conservative reserving assumptions to mitigate the risk of under-reserving in the presence of extreme but plausible losses**

## 6. Schema Mapping & Link Checks

### Expected relationships
- `Claim_ID` links **Claims → Third Parties** (1-to-many possible)
- `Policy_ID` links **Claims → Policyholders** (many-to-1)
- `Accident_Date` links to **Weather** (external enrichment) via date and location/region 

We validate key integrity and join coverage below.


```python
# 6.1 Claim_ID links to Third Parties
claims_claim_ids = set(claims["Claim_ID"].dropna().unique())
tp_claim_ids = set(third_parties["Claim_ID"].dropna().unique())

coverage_tp = len(tp_claim_ids & claims_claim_ids) / max(1, len(tp_claim_ids)) * 100
tp_in_claims = len(tp_claim_ids & claims_claim_ids)

print("Third-party Claim_IDs found in Claims:", tp_in_claims, "/", len(tp_claim_ids), f"({coverage_tp:.2f}%)")
print("Claims with >=1 third party record:", third_parties["Claim_ID"].nunique())

# 6.2 Policy_ID links to Policyholders
claims_policy_ids = set(claims["Policy_ID"].dropna().unique())
ph_policy_ids = set(policyholders["Policy_ID"].dropna().unique())

coverage_ph = len(claims_policy_ids & ph_policy_ids) / max(1, len(claims_policy_ids)) * 100
print("Claims Policy_IDs found in Policyholders:", len(claims_policy_ids & ph_policy_ids), "/", len(claims_policy_ids), f"({coverage_ph:.2f}%)")

# 6.3 Accident_Date links to Weather (readiness check: parseability)
date_cols = ["Accident_Date", "FNOL_Date", "Settlement_Date"]
date_parse_report = []

for c in date_cols:
    parsed = pd.to_datetime(claims[c], errors="coerce")
    invalid = parsed.isna().sum()
    date_parse_report.append({
        "column": c,
        "invalid_or_missing_count": int(invalid),
        "invalid_or_missing_pct": round(float(invalid / len(claims) * 100), 2)
    })

pd.DataFrame(date_parse_report)
```

## **__Insights:__**
**The schema validation confirms strong referential integrity across the core datasets. All third-party claim records successfully link to the claims table, and all policy identifiers in the claims data are present in the policyholder table, indicating complete and reliable relational coverage. This supports robust multi-table joins and reduces the risk of data leakage or record loss during feature engineering.**

**Date field assessment shows complete and valid recording of both Accident_Date and FNOL_Date, enabling accurate temporal analyses such as reporting delays and time-to-settlement modelling. In contrast, Settlement_Date exhibits missing values for approximately 5.3% of claims, which is consistent with claims that remain open or unresolved. This missingness is therefore structural and lifecycle-driven rather than indicative of data quality issues.**

**Overall, the dataset demonstrates high structural quality and operational realism. Missing information is confined to outcome-related fields, reinforcing the appropriateness of training models on closed claims while reserving open claims for prediction and monitoring in a reserving and severity-modelling context.**

## 7. Key Data Quality Issues

This section auto-flags common issues:
- Incomplete/unparseable dates
- Inconsistent categories (case/whitespace)
- High-cardinality categorical fields


```python
# 7.1 Category normalization check 
cat_cols_claims = ["Claim_Type", "Claim_Complexity", "Severity_Band", "Status"]
cat_report = []

for c in cat_cols_claims:
    s = claims[c].astype("string")
    raw_unique = s.nunique(dropna=True)
    norm_unique = s.str.strip().str.lower().nunique(dropna=True)
    cat_report.append({
        "column": c,
        "unique_raw": int(raw_unique),
        "unique_normalized(strip+lower)": int(norm_unique)
    })

display(pd.DataFrame(cat_report))

# Show top categories (useful for spotting inconsistencies)
for c in cat_cols_claims:
    print(f"\nTop values for {c}:")
    display(claims[c].astype("string").value_counts(dropna=False).head(10))

# 7.2 High-cardinality checks
def high_cardinality(df, threshold=50):
    out = []
    for c in df.columns:
        if df[c].dtype == "object" or str(df[c].dtype).startswith("string"):
            n_unique = df[c].nunique(dropna=True)
            if n_unique >= threshold:
                out.append((c, int(n_unique)))
    return pd.DataFrame(out, columns=["column", "unique_values"]).sort_values("unique_values", ascending=False)

print("High-cardinality categorical columns (claims):")
display(high_cardinality(claims, threshold=20))

print("High-cardinality categorical columns (policyholders):")
display(high_cardinality(policyholders, threshold=20))

print("High-cardinality categorical columns (third_parties):")
display(high_cardinality(third_parties, threshold=20))
```

## **__Insights:__**
**The categorical variables in the claims dataset exhibit consistent and well-governed classification structures, with no evidence of label fragmentation due to casing or formatting inconsistencies. The number of raw unique values matches the normalised counts across all key categorical fields, indicating robust data standardisation and reducing the need for corrective preprocessing.**

**Claim Type distribution is dominated by collision-related claims, followed by weather and theft-related events, reflecting a typical personal or commercial motor insurance portfolio. Less frequent categories such as fire and vandalism contribute fewer claims but are likely associated with higher average severity, reinforcing the distinction between frequency-driven and severity-driven risk components.**

**Claim Complexity is heavily skewed toward low-complexity claims, with progressively fewer medium- and high-complexity cases. This distribution aligns with operational expectations, where the majority of claims are straightforward, while a smaller subset requires more intensive handling and is more likely to generate higher costs.**

**Severity Band analysis shows a strong concentration of minor and moderate claims, with a materially smaller proportion of severe and catastrophic losses. Despite their low frequency, severe and catastrophic claims represent a critical severity tail and are expected to drive a disproportionate share of total losses, making them highly relevant for reserving and capital considerations.**

## 8. Research: Missing Value Handling Techniques

**Mean Imputation**
- Fast and simple for numeric fields
- Can distort skewed insurance variables (e.g., claim amounts)

**Median Imputation**
- More robust under skewness and outliers
- Often preferred for monetary insurance fields

**Model-based Imputation**
- Predict missing values using correlated features (e.g., regression / kNN / tree-based)
- More accurate but increases complexity; must avoid leakage (especially with target-related fields)
